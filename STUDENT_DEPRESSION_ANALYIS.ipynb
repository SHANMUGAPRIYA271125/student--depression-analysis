```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Upload and Read File
uploaded = files.upload()
for filename in uploaded.keys():
    data = pd.read_csv(filename, encoding='ISO-8859-1')
```

```python
print(data.head())
```

```python
print(data.info())
```

```python
print(data.describe())
```

```python
#1 Gender-wise Depression Levels
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.boxplot(x= data['Gender'], y= data['Depression'])
plt.title("Depression Levels by Gender")
plt.show()
```

```python
# 2 Work/Study Hours vs Depression
plt.figure(figsize=(8,5))
sns.scatterplot(x=data['Work/Study Hours'], y=data['Depression'])
plt.title("Work/Study Hours vs Depression")
plt.show()
```

```python
# 3. Family History of Mental Illness vs Depression
plt.figure(figsize=(8,5))
sns.boxplot(x=data['Family History of Mental Illness'], y=data['Depression'])
plt.title("Family History of Mental Illness vs Depression")
plt.show()
```

```python
# 4. Suicidal Thoughts vs Depression
plt.figure(figsize=(8,5))
sns.countplot(x=data['Have you ever had suicidal thoughts ?'], hue=data['Depression'])
plt.title("Suicidal Thoughts vs Depression")
plt.show()
```

```python
# 5. Profession vs Depression
plt.figure(figsize=(10,5))
sns.countplot(y=data['Profession'], hue=data['Depression'])
plt.title("Profession vs Depression")
plt.show()
```

```python
# 6 Age vs Depression
plt.figure(figsize=(8,5))
sns.histplot(data=data, x='Age', hue='Depression', bins=15, kde=True)
plt.title("Age vs Depression")
plt.show()
```

```python
# 7. Dietary Habits vs Depression
plt.figure(figsize=(8,5))
sns.boxplot(x=data['Dietary Habits'], y=data['Depression'])
plt.title("Dietary Habits vs Depression")
plt.show()
```

```python
#8. Distribution of depression levels
plt.figure(figsize=(8,5))
sns.countplot(x=data['Depression'])
plt.title("Distribution of Depression Levels")
plt.show()
```

```python
 #9 Academic Pressure vs Depression
plt.figure(figsize=(8,5))
sns.scatterplot(x=data['Academic Pressure'], y=data['Depression'])
plt.title("Academic Pressure vs Depression")
plt.show()
```
